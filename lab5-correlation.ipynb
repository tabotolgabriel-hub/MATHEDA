```python
# import and packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv(r'dataset\current-power.csv',
                 delimiter=",")
df
```

```python
# Summary of Statistics
df.describe()
```

```python
# Covariance of Current and Power
df.cov()
```

```python
# Scatter Plot
x = df['Current']
y = df['Power']

plt.figure()
plt.scatter(x,y)
plt.title('Device Current vs Power Response')
plt.ylabel('Power (W)')
plt.xlabel('Current (A)')
plt.show()
```

```python
df.corr()
```

```python
# regression line
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept

# Scatter Plot
plt.figure(figsize=(7,4))
plt.scatter(x,y,
            color= '#486FB0',
            alpha=0.6,
            label='Data Points')
# line plot
plt.plot(x, line,
         color = "#F46BC6",
         label=f'Regression line (r = {r_value:.3f})')

plt.title('Device Current vs Power Response')
plt.ylabel('Power (W)')
plt.xlabel('Current (A)')
plt.legend()
plt.show()
```

# Air Quality

```python
# load dataset
air = pd.read_csv(r"dataset/air-quality.csv", delimiter=",")
air
```

# NOx-Humidity

```python
#  Regression Line
x = air['NOx']
y = air['Relative_Humidity']
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept

# Scatter Plot
plt.figure(figsize=(7,4))
plt.scatter(x, y,
            color="#1c1f24",
            alpha=0.6,
            label='Data Points')

# Line Plot
plt.plot(x, line,
         color="#FF64DD",
         label=f'Regression line (r = {r_value:.3f})')

plt.title('NOx vs Humidity Correlation')
plt.ylabel('Relative Humidity (%)')
plt.xlabel('NOx (ug/cu.m)')
plt.legend()
plt.show()
```
